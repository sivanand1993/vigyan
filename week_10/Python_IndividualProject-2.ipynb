```python
def female_to_male():
    grouped = merged.groupby(['NOC', 'Sex'])

    count_of_sex_per_country = {}

    for name, group in grouped:
        if name[0] in count_of_sex_per_country.keys():
             count_of_sex_per_country[name[0]].append({'M' : group.count()['ID']})
        else:
            count_of_sex_per_country[name[0]] = []
            count_of_sex_per_country[name[0]].append({'F' : group.count()['ID']})

    female_to_male_ratio = {}
    for country in count_of_sex_per_country:
        if len(count_of_sex_per_country[country]) >= 2:
            female_to_male_ratio[country] = count_of_sex_per_country[country][0]['F']/count_of_sex_per_country[country][1]['M']
   
    sorted_dict = sorted(female_to_male_ratio.items(), key = 
                 lambda kv:(kv[1], kv[0]))

    grouped = merged.groupby(['NOC', 'region'])
    map_noc_country = {}
    
    for name, group in grouped:
        map_noc_country[name[0]] = name[1]
        
    for elem in sorted_dict:
        if elem[0] in map_noc_country.keys():
            print("{:35s}   {:10s}".format(map_noc_country[elem[0]].split('-')[0],   str(elem[1])))
            
female_to_male()
```

```python
#from the above male to female ratio we can understand that for values that are above 1 are the countries that have almost equal male to female athlete ratio.
#Sadly only 6 countries
```

```python
def countries_participated_per_year():
    grouped = merged.groupby(['Year', 'region'])

    count_countries_by_year = {}
    for name, group in grouped:
        if name[0] in count_countries_by_year.keys():
            count_countries_by_year[name[0]] += 1
        else:
            count_countries_by_year[name[0]] = 1

    for year in count_countries_by_year:
            print("{:35s}   {:10s}".format(str(year), str(count_countries_by_year[year])))

countries_participated_per_year()
```

```python
#this code gives us the number of countries participating each year in Olympics. 
#We can see a decline in the middle that is because of winter olympics. 
#The main olympics happen in leap years for which the number has been consistently increasing
```
